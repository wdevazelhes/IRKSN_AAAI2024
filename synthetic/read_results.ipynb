```python
import pandas as pd
import matplotlib.pyplot as plt


import matplotlib as mpl
mpl.rcParams['font.size'] = 26
```

# Plot for n_samples

```python
# Read CSV file
df = pd.read_csv('results_n_samples.csv')
df.index = df['algo']
# # Reshape the dataframe
df_mean = df.pivot(index='algo', columns='n_samples', values='mse')
df_std = df.pivot(index='algo', columns='n_samples', values='std')
```

```python
# Extract algo names
algos = ['lasso', 'enet',  'irksn', 'iht', 'irosr', 'ircr', 'srdi', 'omp', 'ksn']

algos_legend = ['lasso', 'enet',  'irksn', None, None, None, None, None, None]

for i, algo in enumerate(algos):
    mean_scores = df_mean.loc[algo]
    std_scores = df_std.loc[algo]
    n_samples = df_mean.columns.values.astype(int)
    plt.plot(n_samples, mean_scores, label=algos_legend[i], marker='o')
    plt.fill_between(n_samples, mean_scores - std_scores, mean_scores + std_scores, alpha=0.2, edgecolor='face', linewidth=0.5)

plt.legend(loc='upper left', handlelength=0.5, ncol=1, labelspacing=0.001, framealpha=0, mode='expand', columnspacing=0.001)
# plt.legend(loc='lower center', handlelength=0.3, ncol=3, labelspacing=0.001)

# Set the labelspacing to a smaller value (adjust as needed)
# legend.get_title().set_pad(0.1)#
# legend.get_frame().set_alpha(0.1)

# plt.legend(['lasso', 'iht', 'srdi', 'enet'], handlelength=0.3, ncol=3, loc='lower center')




plt.grid(False)
plt.show()
```
